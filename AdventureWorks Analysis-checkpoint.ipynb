```python
unique_products = prod_conn['ProductSize'].unique()

# Print the unique values
print(unique_products)
```

```python
# to convert the numbers to sizes, the folloing steps were done
# minimum value - max value = 38 - 70 = 32
# 32 / 4 (representing the 4 different sizes) = 8
# 8 will be added to each value to create a range

# Define a function to map sizes
def product_size(x):
    try:
        # Convert to integer if possible
        x = int(x)
    except ValueError:
        # Handle non-integer values (e.g., empty strings or non-numeric data)
        return "Unknown"

    # Now apply the size mapping logic
    if x >= 62:
        return "XL"
    elif x >= 54:
        return "L"
    elif x >= 46:
        return "M"
    elif x >= 38:
        return "S"
    else:
        return "Unknown"  # For values outside the expected range i.e 0 
```

```python
# Apply the function to the ProductSize column in prod_conn DataFrame

prod_conn['ProductSize'] = prod_conn['ProductSize'].apply(product_size)
```

```python
# To view the modified DataFrame or view the changes
print(prod_conn['ProductSize'])
```

## Creating relationships between the tables

```python
# Join the Sales table and Product's table using the 'ProductKey' column
salesprod = pd.merge(sales_conn, prod_conn, on='ProductKey')

salesprod
```

```python
# Join the Calender table and Sales's table
calendsales = pd.merge(calend_conn, sales_conn, left_on='Date', right_on='OrderDate', how='left')

calendsales
```

```python
# Joining the Calender, Sales and Product's table using the 'ProductKey' column
calensalesprod = pd.merge(calendsales, prod_conn, on='ProductKey')


calensalesprod
```

```python
# Join the Calender, Sales and Product's table to Territory table using the 'TerritoryKey' column
calensalesprodterr = pd.merge(calensalesprod,territory_conn, left_on='TerritoryKey', right_on='SalesTerritoryKey', how='left')

calensalesprodterr
```

```python
# Join the Calender, Sales, Product and Territory table to customer's table using the 'CustomerKey' column
calsalesprodterrcust = pd.merge(calensalesprodterr,cust_connect, left_on='CustomerKey', right_on='CustomerKey', how='left')

calsalesprodterrcust
```
